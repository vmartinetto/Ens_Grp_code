```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
```

```python
evecs = np.loadtxt('eigenvectors.txt',usecols=(0,1))
evals = np.loadtxt('eigenvalues.txt')
```

```python
N = int(np.sqrt(evecs.shape[0]))
psi0 = evecs[:,0]
psi0mat = np.reshape(psi0,(N,N))
dens0mat = psi0mat**2
```

```python
Nsites = 100
boxsize = 1.0
dx = boxsize/Nsites
xx = np.linspace(0,1,Nsites)
#xx = dx*(range(Nsites)-float(Nsites-1)/2*np.ones(Nsites))

eval = evals[0]
den = np.empty(N)
psi = np.empty(N)
for i in range(N):
    den[i] = integrate.simps(dens0mat[i,:],xx)
    psi[i] = integrate.simps(psi0mat[i,:],xx)
plt.plot(xx,psi)
plt.show()
plt.plot(xx,den)
plt.show()

dens = psi**2
plt.plot(xx,dens)
plt.show()
norm = integrate.simps(dens,xx)
print(norm)
```

```python
den = np.sum(dens0mat,axis=1)
norm = np.sum(den)
print(norm)
```
